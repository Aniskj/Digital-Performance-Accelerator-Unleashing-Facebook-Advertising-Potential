We can perform feature engineering using the available dataset to create key performance indicators (KPIs) that are crucial for subsequent analysis:

1. Conversion Rate (CR):
   - CR is a metric that quantifies the proportion of users who express interest in a product by making an inquiry after viewing an advertisement. It enables us to gauge the percentage of users who were motivated by the ad during a marketing campaign and actually made a purchase.

2. Click-Through Rate (CTR):
   - CTR represents the ratio of users who clicked on an ad to the total number of users who viewed that ad. In simpler terms, CTR measures the percentage of users who found the ad engaging and relevant enough to take action by clicking on it. This KPI helps assess how effective the ads are in capturing the attention of the intended audience.

3. Cost Per Click (CPC):
   - CPC indicates the expense incurred for each click generated by an ad. This metric aids in evaluating the cost-efficiency of acquiring an additional user's attention through the ad. It tells us whether the spending on attracting more users to engage with the ad is economical or not.

# A/B TESTING

To obtain a more accurate evaluation of the actual performance of the campaigns, we need to examine the selected Key Performance Indicators (KPIs) - CR, CPC, and CTR, as mentioned earlier. In brief, CR (conversion rate) gauges the conversion effectiveness of each campaign, CPC (cost per click) assesses the cost-efficiency of each click an ad set receives, and CTR (click-through rate) indicates the ability to attract and engage customers. These metrics are expressed as ratios, which enables a more robust and precise comparison between advertising campaigns, even when there are significant differences in spending.

To determine whether the differences between these ad campaigns are statistically significant or merely due to chance, we can conduct A/B testing. The primary goal of A/B testing for these three groups is to perform pairwise comparisons between each pair of groups (A-B, A-C, and B-C) to ascertain whether the differences between them are statistically significant or not. 

A/B testing is used to determine if differences between ad campaigns are statistically significant or just due to chance. When conducting A/B testing with three groups, the main goal is to compare each pair of groups (A-B, A-C, and B-C) to determine if the differences between them are statistically significant. However, when you make multiple comparisons simultaneously, the risk of making false positive conclusions (Type I error rate) increases. This issue is known as the multiple comparison problem, where making several inferences at once can lead to incorrect interpretation of results. To address this problem, you can employ the Bonferroni correction as a post-hoc method to make the significance threshold stricter to account for the multiple inferences being made. Here are the detailed steps for implementing the Bonferroni correction:

To assess the normality of group residuals, use the Shapiro-Wilk test. If the residuals follow a normal distribution, then proceed to examine the homogeneity of variances among the groups using Levene's test. If the variances are homogeneous, carry out pairwise T-tests. However, if variance homogeneity is not met, conduct pairwise Welch's tests. In cases where the residuals are not normally distributed, employ pairwise Mann-Whitney U tests. Finally, to draw a conclusive result, apply the Bonferroni correction.

Note:
To establish the significance level, we typically utilize the standard α value of 0.05. To apply the Bonferroni correction, we adjust the significance threshold by dividing the initial α value of 0.05 by the number of pairwise comparisons, which in this case is 3. The adjusted significance threshold becomes:α /3=0.05/3=0.01666.

# Step 1:Check if the residuals are normally distributed

Null Hypothesis(H0):Sample distribution are close to normal

Alternate Hypothesis:Sample distribution are not close to normal.

Therefore:

If p-value < 0.05, H0 = Reject

If p-value > 0.05 H0 = Cannot Reject.

```python
#For Conversion Rate(CR)
pvalue1 = shapiro(df.loc[df["xyz_campaign_id"] == "campaign_1", "CR"])
pvalue2 = shapiro(df.loc[df["xyz_campaign_id"] == "campaign_2", "CR"])
pvalue3 = shapiro(df.loc[df["xyz_campaign_id"] == "campaign_3", "CR"])

print("p-value for campaign_1: ", pvalue1)
print("p-value for campaign_2: ", pvalue2)
print("p-value for campaign_3: ", pvalue3)
```

Here we can see that all 3 campaigns yield p value much lower than 0.05,therefore we reject null hypothesis.Therefore the group residuals are not normal.

```python
#For Click Through Rate(CTR)
pvalue1 = shapiro(df.loc[df["xyz_campaign_id"] == "campaign_1", "CTR"])
pvalue2 = shapiro(df.loc[df["xyz_campaign_id"] == "campaign_2", "CTR"])
pvalue3 = shapiro(df.loc[df["xyz_campaign_id"] == "campaign_3", "CTR"])

print("p-value for campaign_1: ", pvalue1)
print("p-value for campaign_2: ", pvalue2)
print("p-value for campaign_3: ", pvalue3)
```

Here we can see that all 3 campaigns yield p value much lower than 0.05,therefore we reject null hypothesis.Therefore the group residuals are not normal.

```python
#For Cost per Click(CPC)
pvalue1 = shapiro(df.loc[df["xyz_campaign_id"] == "campaign_1", "CPC"])
pvalue2 = shapiro(df.loc[df["xyz_campaign_id"] == "campaign_2", "CPC"])
pvalue3 = shapiro(df.loc[df["xyz_campaign_id"] == "campaign_3", "CPC"])

print("p-value for campaign_1: ", pvalue1)
print("p-value for campaign_2: ", pvalue2)
print("p-value for campaign_3: ", pvalue3)
```

Here we can see that all 3 campaigns yield p value much lower than 0.05,therefore we reject null hypothesis.Therefore the group residuals are not normal.

# Step 2: Check for Variance Homogenity

H0: Variance is homogeneous.

H1: Variance is not homogeneous.


If p-value < 0.05, H0 = Reject

If p-value > 0.05 H0 = Cannot Reject.

```python
#For Conversion Rate(CR)
pvalue1 = levene(df.loc[df["xyz_campaign_id"] == "campaign_1", "CR"],
                df.loc[df["xyz_campaign_id"] == "campaign_2", "CR"],
                df.loc[df["xyz_campaign_id"] == "campaign_3", "CR"])

print("p-value for Levene's test: ", pvalue1)
```

```python
#For Click Through Rate(CTR)
pvalue2 = levene(df.loc[df["xyz_campaign_id"] == "campaign_1", "CTR"],
                df.loc[df["xyz_campaign_id"] == "campaign_2", "CTR"],
                df.loc[df["xyz_campaign_id"] == "campaign_3", "CTR"])

print("p-value for Levene's test: ", pvalue2)

```

```python
#For Cost Per Click(CPC)
pvalue3 = levene(df.loc[df["xyz_campaign_id"] == "campaign_1", "CPC"],
                df.loc[df["xyz_campaign_id"] == "campaign_2", "CPC"],
                df.loc[df["xyz_campaign_id"] == "campaign_3", "CPC"])

print("p-value for Levene's test: ", pvalue3)
```

As we have seen above for all three cases the pvalue<0.05.Therefore we reject null hypothesis.Therefore we acn say that Variance is not Homogenous.

Since all three groups of all three KPIs fail to satisfy the normality condition and variance homogenity test, we can use the non-parametric Mann Whitney U test.


# Step 3:Apply Mann Whitney U Test

Null hypothesis: There is no difference (in terms of central tendency) between the two groups in the population

Alternate hypothesis: There is a difference (with respect to the central tendency) between the two groups in the population.

```python
#For Conversion Rate(CR)
pvalue1 = mannwhitneyu(df.loc[df["xyz_campaign_id"] == "campaign_1", "CR"], df.loc[df["xyz_campaign_id"] == "campaign_2", "CR"])
pvalue2 = mannwhitneyu(df.loc[df["xyz_campaign_id"] == "campaign_2", "CR"], df.loc[df["xyz_campaign_id"] == "campaign_3", "CR"])
pvalue3 = mannwhitneyu(df.loc[df["xyz_campaign_id"] == "campaign_3", "CR"], df.loc[df["xyz_campaign_id"] == "campaign_1", "CR"])

print("p-value for campaign_1-campaign_2: ", pvalue1)
print("p-value for campaign_2-campaign_3: ", pvalue2)
print("p-value for campaign_3-campaign_1: ", pvalue3)
```

```python
#For Click Through Rate(CTR)
pvalue1 = mannwhitneyu(df.loc[df["xyz_campaign_id"] == "campaign_1", "CTR"], df.loc[df["xyz_campaign_id"] == "campaign_2", "CTR"])
pvalue2 = mannwhitneyu(df.loc[df["xyz_campaign_id"] == "campaign_2", "CTR"], df.loc[df["xyz_campaign_id"] == "campaign_3", "CTR"])
pvalue3 = mannwhitneyu(df.loc[df["xyz_campaign_id"] == "campaign_3", "CTR"], df.loc[df["xyz_campaign_id"] == "campaign_1", "CTR"])

print("p-value for campaign_1-campaign_2: ", pvalue1)
print("p-value for campaign_2-campaign_3: ", pvalue2)
print("p-value for campaign_3-campaign_1: ", pvalue3)
```

```python
#For Cost Per Click(CPC)
pvalue1 = mannwhitneyu(df.loc[df["xyz_campaign_id"] == "campaign_1", "CPC"], df.loc[df["xyz_campaign_id"] == "campaign_2", "CPC"])
pvalue2 = mannwhitneyu(df.loc[df["xyz_campaign_id"] == "campaign_2", "CPC"], df.loc[df["xyz_campaign_id"] == "campaign_3", "CPC"])
pvalue3 = mannwhitneyu(df.loc[df["xyz_campaign_id"] == "campaign_3", "CPC"], df.loc[df["xyz_campaign_id"] == "campaign_1", "CPC"])

print("p-value for campaign_1-campaign_2: ", pvalue1)
print("p-value for campaign_2-campaign_3: ", pvalue2)
print("p-value for campaign_3-campaign_1: ", pvalue3)
```

In all three cases, when compared to the adjusted significance level of 0.01667, we find that there is no significant difference between campaigns_1 and campaign_2. However, we do observe a statistically significant difference between campaigns_2 and campaign_3, as well as between campaigns 3 and 1. This indicates that the variations in Conversion Rate (CR), Click-Through Rate (CTR), and Cost Per Click (CPC) between campaigns 2 and 3, and campaigns 3 and 1 are indeed significant and needs attention.

For the company, this implies that making changes in ad campaigns 1 and 2 is unlikely to have a significantly different impact on CR, CTR, and CPC compared to each other. In other words, their relative performance, as measured by these KPIs, is quite similar. On the contrary, we can detect notable differences between campaigns 2 and 1 in comparison to campaign 3. This suggests that, relative to their respective spending levels, campaigns 2 and 1 have performed significantly better in terms of conversion, engagement, and cost-efficiency than campaign 3, which received a much higher investment.
